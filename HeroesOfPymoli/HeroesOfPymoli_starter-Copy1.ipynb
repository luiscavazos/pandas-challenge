```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
conteo = len(purchase_data["SN"].unique())
P_count = pd.DataFrame({"Total Players" : [conteo]})
P_count
```

## Purchasing Analysis (Total)

```python
Unique_num = len(purchase_data["Item ID"].unique())
price_avg = purchase_data["Price"].mean()
purchase_num = purchase_data["Purchase ID"].count()
Revenue = purchase_data["Price"].sum()
Purchase_Analysis = pd.DataFrame({"Number of Unique Items" : [Unique_num], "Average Price" : [price_avg], 
                                 "Number of Purchases" : [purchase_num], "Total Revenue" : [Revenue]})
Purchase_Analysis["Average Price"] = Purchase_Analysis["Average Price"].astype(float).map("${:,.2f}".format)
Purchase_Analysis["Total Revenue"] = Purchase_Analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
Purchase_Analysis.head()
```

```python
gender_group = purchase_data.groupby(["Gender"])
gender_group = gender_group.nunique()
t_gender = gender_group["SN"].sum()
count =gender_group["SN"].unique()
percent = (gender_group["SN"]/ t_gender) *100
gender_demo = pd.DataFrame({"Total Count" : count, "Percentage of Players" : percent})
gender_demo["Percentage of Players"] = gender_demo["Percentage of Players"].astype(float).map("{:,.2f}%".format)
gender_demo
```

```python
pur_ana = purchase_data.groupby(["Gender"])

count = pur_ana["SN"].count()
avg_price = pur_ana["Price"].mean()
suma = pur_ana["Price"].sum()
avg_person = avg_price/count #??? no entiendo que es el AVG Total purchase per person debe dar $4.47, 4.07 y 4.56
purchase_anal = pd.DataFrame({"Purchase Count" : count, "Average Purchase Price" : avg_price, "Total Purchase Value" : suma, 
                              "Avg Total Purchase per Person": avg_person})
purchase_anal["Average Purchase Price"] = purchase_anal["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchase_anal["Total Purchase Value"] = purchase_anal["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchase_anal["Avg Total Purchase per Person"] = purchase_anal["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)


purchase_anal
```

```python
bins = [0,10, 14, 19, 24, 29,34, 39, 40]
group_names= ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39",">40"]
age_df = purchase_data
age_df["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
age_df_final = age_df.groupby("Age Group")
age_df_final
```

```python
countage = age_df_final["SN"].count()
avg_priceage = age_df_final["Price"].mean()
sumage = age_df_final["Price"].sum()
avg_personage = avg_priceage/countage #??? no entiendo que es el AVG Total purchase per person debe dar $4.47, 4.07 y 4.56
Analysis_age = pd.DataFrame({"Purchase Count" : countage, "Average Purchase Price" : avg_priceage, "Total Purchase Value" : sumage, 
                              "Avg Total Purchase per Person": avg_personage})
Analysis_age["Average Purchase Price"] = Analysis_age["Average Purchase Price"].astype(float).map("${:,.2f}".format)
Analysis_age["Total Purchase Value"] = Analysis_age["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Analysis_age["Avg Total Purchase per Person"] = Analysis_age["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

Analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders = purchase_data.groupby(["SN"])
spnd_count= spenders.count()["Price"]
spnd_avg= spenders.mean()["Price"]
spnd_tot= spenders.sum()["Price"]
spend_analysis = pd.DataFrame({"Purchase Count" : spnd_count, "Average Purchase Price" : spnd_avg, 
                               "Total Purchase Value" :spnd_tot})
spend_analysis["Average Purchase Price"] = spend_analysis["Average Purchase Price"].astype(float).map("${:,.2f}".format)
spend_analysis["Total Purchase Value"] = spend_analysis["Total Purchase Value"].astype(float).map("${:,.2f}".format)

Spend_final =spend_analysis.sort_values("Total Purchase Value", ascending=False)
Spend_final.head(5)
```

## Most Popular Items

```python
Popular = purchase_data[["Item ID", "Item Name", "Price"]]
Pop= Popular.groupby(["Item ID", "Item Name"])
Pop_count= Pop["Item ID"].count()
Pop_price= Pop["Price"].mean()
Pop_tot = Pop["Price"].sum()
Pop_analysis = pd.DataFrame({"Purchase Count" : Pop_count, "Item Price" : Pop_price, "Total Purchase Value" : Pop_tot})
Pop_final =Pop_analysis.sort_values("Total Purchase Value", ascending=False)
Pop_final.head(5)
```
